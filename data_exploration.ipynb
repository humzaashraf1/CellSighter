### Data: ./cells
A folder of .npz files that contain label images of segmentation. Below we can just display the label image for the first sample.

```python
data = np.load('./example_experiment/MelanomaLymphNode/cells/1.npz')

cells = data['data']
cells = cells.astype(float)
print(type(cells))

unique_labels = np.unique(cells)
colors = np.random.rand(len(unique_labels) - 1, 3)
background_color = np.array([0, 0, 0])
colors = np.vstack([background_color, colors])
cmap = mcolors.ListedColormap(colors)

plt.imshow(cells, cmap=cmap, interpolation='nearest')
plt.axis('off')
plt.show()

print(cells.shape)
print(unique_labels.shape)
```

### Data: ./data/images
A folder of .npz files that contain the normalized, multiplexed image data. Below we can just display one of the channels (E.g., CD14) for the first sample.

```python
data = np.load('./example_experiment/MelanomaLymphNode/data/images/1.npz')

images = data['data']
print(images.shape)

channel = 3 # E.g., this is CD14
image_slice = images[:,:,channel]

plt.imshow(images[:,:,channel], cmap='gray', vmin=0, vmax=1)
plt.axis('off')
plt.show()
```
